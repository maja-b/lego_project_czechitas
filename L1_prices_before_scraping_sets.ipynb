```python
import pandas as pd
prices = pd.read_csv('current_prices.csv') #Lego prices from Kaggle
age = pd.read_csv('retail_prices_age.csv') #Maven dataset
prices_all = pd.merge(prices,age, how = 'outer', left_on = 'Set_ID', right_on = 'set_id')
# checking for differences between the retail prices of the two datasets
problem = prices_all[(prices_all['USD_MSRP'] != prices_all['US_retailPrice']) & (prices_all['USD_MSRP'].notnull() & prices_all['US_retailPrice'].notnull())] #finds sets where the retail price data are different --> found nothing, great
problem[['Set_ID','set_id','USD_MSRP','US_retailPrice']]
```

```python
problem = prices_all[(prices_all['Name'] != prices_all['name']) & (prices_all['Name'].notnull() & prices_all['name'].notnull())] #finds different set names - there aren't many and the differences aren't big
problem[['Set_ID','set_id','Name','name']]
```

```python
prices_all.info()
```

```python
sets = pd.read_csv('sets.csv')
prices_sets = pd.merge(problem,sets, left_on = 'Set_ID', right_on = 'set_num')
prices_sets[['Name','name_x','name_y']] #name comparison with the set table - sometimes a different name in all 3 tables
```

```python
sets.info()
```

```python
problem = prices_all[(prices_all['Pieces'] != prices_all['pieces']) & (prices_all['Pieces'].notnull() & prices_all['pieces'].notnull())]
problem[['Set_ID','set_id','Pieces','pieces']]
pieces = pd.merge(problem, sets, left_on = 'Set_ID', right_on = 'set_num')
pieces[['set_num','Pieces','pieces','num_parts']] #comparison of the number of Lego pieces in cases where Maven and Kaggle datasets differ, with the sets table - again 3 different values
```

```python
prices_sets = pd.merge(prices,sets, how = 'outer', left_on = 'Set_ID', right_on = 'set_num')
problem = prices_sets[(prices_sets['Pieces'] != prices_sets['num_parts']) & (prices_sets['Pieces'].notnull() & prices_sets['num_parts'].notnull())]
problem[['set_num','Pieces','num_parts']] #merge of current_prices and sets table - in 4777 cases the number of Lego pieces differs
```

```python
problem = prices_sets[(prices_sets['year'] != prices_sets['Year']) & (prices_sets['year'].notnull() & prices_sets['Year'].notnull())]
problem['yeardif'] = problem['Year']-problem['year']
problem[['Set_ID','Year','year','yeardif']].sort_values('yeardif') #comparison of release years between current_prices (Kaggle) and sets (Rebrickable) - 489 sets have a difference, the largest difference is 23 years - for older sets it is not possible to find reliable information about the release year
```

```python
prices_sets['set_id']=prices_sets[['set_num','Set_ID']].bfill(axis=1).iloc[:, 0] #takes the first non-null set ID value from the sets and current_prices tables and creates a new column
prices_sets.drop(columns = ['set_num','Set_ID'],inplace = True)
```

```python
prices_sets[prices_sets['Name'].notnull() & prices_sets['name'].isnull()].info()
```

```python
prices_sets[prices_sets['Name'].isnull() & prices_sets['name'].notnull()].info() #9399 sets where one of the names is null
```

```python
prices_sets[(prices_sets['Name'].notnull() & prices_sets['name'].notnull())].info() #14062 where there is no null - after merging the columns I expect 23461 non-null set names
```

```python
prices_sets['set_name']=prices_sets[['name','Name']].bfill(axis=1).iloc[:, 0]
prices_sets.drop(columns = ['name','Name'],inplace = True)
prices_sets.info() #after merging the columns I really have 23461 non-null values ​​in the set name, which are all the rows
```

```python
prices_sets['set_year']=prices_sets[['year','Year']].bfill(axis=1).iloc[:, 0]
prices_sets.drop(columns = ['year','Year'],inplace = True)
prices_sets.info() #the year is now filled everywhere and is in 1 column
```

```python
prices_sets['set_pieces']=prices_sets[['num_parts','Pieces']].bfill(axis=1).iloc[:, 0] #we primarily want to keep the data from Rebrickable
prices_sets.drop(columns = ['num_parts','Pieces'],inplace = True)
prices_sets.info() #the number of pieces for some sets (about 500) is missing; the table is now free of duplicate columns and is ready for the next merge
```

```python
prices_sets_age = pd.merge(prices_sets,age, how = 'outer', on = 'set_id') #merged table with retail prices and age categories
prices_sets_age.info()
```

```python
problem = prices_sets_age[(prices_sets_age['Minifigures'] != prices_sets_age['minifigs']) & (prices_sets_age['Minifigures'].notnull() & prices_sets_age['minifigs'].notnull())]
problem[['set_id','Minifigures','minifigs']] #there aren't many differences and it's always one minifigure, we'll keep the data from Rebrickable because they seem to be more accurate according to Google
```
