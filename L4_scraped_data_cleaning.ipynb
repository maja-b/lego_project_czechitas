```python
import pandas as pd
brickset = pd.read_csv('brickset.csv') #scraped data from Brickset
brickset.info()
```

```python
brickset.drop(columns = ['Notes','Wanted'],inplace = True) #these columns only contain null values so I don't need them
```

```python
brickset[brickset['RRP (USD)'].isnull() & (brickset['RRP (GBP)'].notnull() | brickset['RRP (CAD)'].notnull() | brickset['RRP (EUR)'].notnull())] #there are rows with missing USD retail price which have prices in other currencies so I will need to convert those
```

```python
brickset[brickset['RRP (USD)'].isnull() & brickset['RRP (CAD)'].notnull()].info()
```

```python
brickset[brickset['RRP (USD)'].isnull() & brickset['RRP (EUR)'].notnull()].info()
```

```python
brickset[brickset['RRP (USD)'].isnull() & brickset['RRP (GBP)'].notnull()].info()
```

```python
brickset['RRP (USD from GBP)'] = brickset['RRP (GBP)']*1.25 #FX for the day of scraping from Google
brickset['RRP (USD)'] = brickset[['RRP (USD)','RRP (USD from GBP)']].bfill(axis=1).iloc[:, 0] #keeping primarily the original USD value and only replacing nulls
brickset.drop(columns='RRP (USD from GBP)',inplace=True)
brickset.info()
```

```python
brickset['RRP (USD from EUR)'] = brickset['RRP (EUR)']*1.08 #FX for the day of scraping from Google
brickset['RRP (USD)'] = brickset[['RRP (USD)','RRP (USD from EUR)']].bfill(axis=1).iloc[:, 0]
brickset.drop(columns='RRP (USD from EUR)',inplace=True)
brickset.info()
```

```python
brickset[brickset['RRP (USD)'].isnull() & (brickset['RRP (GBP)'].notnull() | brickset['RRP (CAD)'].notnull() | brickset['RRP (EUR)'].notnull())] #we have all the prices in USD now, no need to convert CAD
```

```python
brickset.drop(columns=['RRP (GBP)','RRP (CAD)','RRP (EUR)'],inplace=True)
brickset.info()
```

```python
bricklink = pd.read_csv('bricklink.csv')
bricklink.info()
```

```python
bricklink.drop(columns=['last_6_months_new_total_lots','last_6_months_new_min_price','last_6_months_new_avg_price','last_6_months_new_max_price','current_new_total_lots','current_new_min_price','current_new_max_price','current_new_avg_price','last_6_months_used_total_lots','last_6_months_used_min_price','last_6_months_used_avg_price','last_6_months_used_max_price','current_used_total_lots','current_used_min_price','current_used_avg_price','current_used_max_price'],inplace=True)
bricklink.info() #I got rid of columns which we will not need for our analyses
```

```python
#data cleaning to convert the prices into floats instead of strings
bricklink['last_6_months_new_qty_avg_price'] = bricklink['last_6_months_new_qty_avg_price'].str.strip('USÂ $')
bricklink['last_6_months_new_qty_avg_price'] = bricklink['last_6_months_new_qty_avg_price'].str.replace(',','').astype(float)
bricklink['last_6_months_used_qty_avg_price'] = bricklink['last_6_months_used_qty_avg_price'].str.strip('USÂ $')
bricklink['last_6_months_used_qty_avg_price'] = bricklink['last_6_months_used_qty_avg_price'].str.replace(',','').astype(float)
bricklink['current_new_qty_avg_price'] = bricklink['current_new_qty_avg_price'].str.strip('USÂ $')
bricklink['current_new_qty_avg_price'] = bricklink['current_new_qty_avg_price'].str.replace(',','').astype(float)
bricklink['current_used_qty_avg_price'] = bricklink['current_used_qty_avg_price'].str.strip('USÂ $')
bricklink['current_used_qty_avg_price'] = bricklink['current_used_qty_avg_price'].str.replace(',','').astype(float)
bricklink.info()
```

```python
bsbl_data = pd.merge(brickset,bricklink,how='outer',left_on = 'Number', right_on = 'set_num')
bsbl_data[bsbl_data['Number'].isnull()]
```

```python
bsbl_data.drop(columns = 'set_num',inplace=True)
bsbl_data.rename(columns = {'Number':'bsbl_id','Theme':'bs_theme','Subtheme':'bs_subtheme','Year':'bs_year','Set name':'bs_name','Minifigs':'bs_minifigures','Pieces':'bs_pieces','RRP (USD)':'bs_rrp','Qty owned':'bs_owned','Qty owned new':'bs_owned_new','Qty owned used':'bs_owned_used','Qty wanted':'bs_wanted','Value new (USD)':'bs_value_new','Value used (USD)':'bs_value_used','last_6_months_new_total_qty':'bl_sold_qty_new','last_6_months_new_qty_avg_price':'bl_sold_avgprice_new','last_6_months_used_total_qty':'bl_sold_qty_used','last_6_months_used_qty_avg_price':'bl_sold_avgprice_used','current_new_total_qty':'bl_offer_qty_new','current_used_total_qty':'bl_offer_qty_used','current_new_qty_avg_price':'bl_offer_avgprice_new','current_used_qty_avg_price':'bl_offer_avgprice_used'},inplace=True)
bsbl_data.drop(columns = ['EAN','UPC','Width','Height','Depth','Weight','Priority','Launch date','Exit date'],inplace=True)
bsbl_data.info()
```

```python
bsbl_data.to_csv('bsbl_data.csv',index=False)
```
