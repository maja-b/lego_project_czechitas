```python
import pandas as pd
import numpy as np
bsbl = pd.read_csv('bsbl_data.csv')
bsbl.info()
```

```python
old = pd.read_csv('prices_sets_age_themesE.csv')
big_merge = pd.merge(old, bsbl, how = 'outer', left_on = 'set_id',right_on='bsbl_id') #merging with the dataset containing "old" prices
big_merge.info()
```

```python
big_merge['set_id']=big_merge[['set_id','bsbl_id']].bfill(axis=1).iloc[:, 0] #merging columns into 1, primarily keeping Rebrickable data (unless indicated otherwise)
big_merge['name']=big_merge[['name','bs_name']].bfill(axis=1).iloc[:, 0]
big_merge['set_theme']=big_merge[['set_theme','bs_theme']].bfill(axis=1).iloc[:, 0]
big_merge['set_subtheme']=big_merge[['set_subtheme','bs_subtheme']].bfill(axis=1).iloc[:, 0]
big_merge['year']=big_merge[['year','bs_year']].bfill(axis=1).iloc[:, 0]
big_merge['minifigures']=big_merge[['minifigures','bs_minifigures']].bfill(axis=1).iloc[:, 0]
big_merge['num_pieces']=big_merge[['num_pieces','bs_pieces']].bfill(axis=1).iloc[:, 0]
big_merge.drop(columns = ['bsbl_id','bs_name','bs_theme','bs_subtheme','bs_year','bs_minifigures','bs_pieces'],inplace = True)
```

```python
problem = big_merge[(big_merge['bs_rrp'] != big_merge['retail_price']) & big_merge['bs_rrp'].notnull() & big_merge['retail_price'].notnull()]
problem['price-dif'] = problem['retail_price']-problem['bs_rrp']
problem[['set_id','bs_rrp','retail_price','price-dif']].sort_values('price-dif') #checking for the differences in the retail prices from the two data sources, there are only a few
```

```python
big_merge['retail_price']=big_merge[['bs_rrp','retail_price']].bfill(axis=1).iloc[:, 0] #in this case I am keeping primarily the data from the scraped websites as they are newer
big_merge.drop(columns = 'bs_rrp',inplace = True)
```

```python
problem = big_merge[(big_merge['bs_owned'] != big_merge['Owned']) & big_merge['bs_owned'].notnull() & big_merge['Owned'].notnull() & big_merge['bs_owned'] != 0]
problem[['set_id','bs_owned','Owned']] #no differences found
```

```python
big_merge['Owned']=big_merge[['Owned','bs_owned']].bfill(axis=1).iloc[:, 0]
big_merge.drop(columns = 'bs_owned',inplace = True)
big_merge.info()
```

```python
big_merge.drop(columns = ['percent_price_dif','Total_Quantity','Current_Price'],inplace = True) #these are old prices columns which we cannot use now
```

```python
# replacing zeroes with nulls where zeroes do not make logical sense
big_merge['retail_price'] = big_merge['retail_price'].replace(0,np.nan)
big_merge['Owned'] = big_merge['Owned'].replace(0,np.nan)
big_merge['Rating'] = big_merge['Rating'].replace(0,np.nan)
big_merge['bs_owned_new'] = big_merge['bs_owned_new'].replace(0,np.nan)
big_merge['bs_owned_used'] = big_merge['bs_owned_used'].replace(0,np.nan)
```

```python
#calculations to be used for our analyses and visualizations later
big_merge['bl_percent_price_increase_new'] = big_merge['bl_sold_avgprice_new']/big_merge['retail_price']-1
big_merge['bl_percent_price_increase_used'] = big_merge['bl_sold_avgprice_used']/big_merge['retail_price']-1
big_merge['bl_price_increase_new'] = big_merge['bl_sold_avgprice_new'] - big_merge['retail_price']
big_merge['bl_price_increase_used'] = big_merge['bl_sold_avgprice_used'] - big_merge['retail_price']
```

```python
#additional data cleaning for themes to unify some names which only differ in spelling, etc.
big_merge['set_theme'] = big_merge['set_theme'].replace('Collectable minifigures','Collectable Minifigures')
big_merge['set_theme'] = big_merge['set_theme'].replace('Creator','Creator Expert')
big_merge['set_theme'] = big_merge['set_theme'].replace('DUPLO','Duplo')
big_merge['set_theme'] = big_merge['set_theme'].replace('Educational','Education')
big_merge['set_theme'] = big_merge['set_theme'].replace('LEGOLAND','Legoland')
big_merge['set_theme'] = big_merge['set_theme'].replace('Promotional ','Promotional')
big_merge['set_theme_group'] = big_merge['set_theme_group'].replace('Constraction','Construction')
big_merge['r_theme'] = big_merge['r_theme'].replace('Constraction','Construction')
```

```python
big_merge.info()
```

```python
big_merge.drop(columns=['bs_owned_new','bs_owned_used'],inplace=True) #these only contain null values after replacing zeroes
```

```python
big_merge.to_csv('big_merge.csv',index=False)
```
